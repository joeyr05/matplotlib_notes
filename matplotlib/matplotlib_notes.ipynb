### Necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#long hand notation
#plt.plot(x,y,label='2x', color='grey', linewidth = 2, linestyle = '--', marker='.', markersize=12, markeredgecolor = 'black')

#short hand notation
#line one
plt.plot(x,y,'k.--',label='2x')

#line two
x2 = np.arange(0,4,0.5)
plt.plot(x2,x2**2,'r',label='x^2')

plt.title("Our First Graph!", fontdict={"fontname":"Times new Roman", "fontsize":18})
plt.xlabel("X-axis", fontdict={"fontname":"Times new Roman"})
plt.ylabel("Y-axis", fontdict={"fontname":"Times new Roman"})

#ticks change the size of the graph concerning the values
plt.xticks([1,2,3,4,5])
#plt.yticks([2,4,6,8,10])

plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#long hand notation
#plt.plot(x,y,label='2x', color='grey', linewidth = 2, linestyle = '--', marker='.', markersize=12, markeredgecolor = 'black')

# Resizing the graph
plt.figure(figsize=(5,3),dpi=90)
#short hand notation
#line one
plt.plot(x,y,'k.--',label='2x')

#line two
x2 = np.arange(0,4,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.title("Our First Graph!", fontdict={"fontname":"Times new Roman", "fontsize":18})
plt.xlabel("X-axis", fontdict={"fontname":"Times new Roman"})
plt.ylabel("Y-axis", fontdict={"fontname":"Times new Roman"})

#ticks change the size of the graph concerning the values
plt.xticks([1,2,3,4,5])
#plt.yticks([2,4,6,8,10])

plt.legend()

plt.savefig('graph.png',dpi=100)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop())

#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))


plt.show()
```
